# Lectura y escritura de autómatas con pila en ficheros 

Los autómatas con pila los creamos leyéndolos desde ficheros. El formato de un fichero con un autómata con pila será el siguiente:

- La primera línea contiene los estados del autómata. Esa línea siempre empezará por "Q = ", para indicar que se está especificando el conjunto de estados. Tal conjunto vendrá comprendido entre llaves. Los estados irán separados por comas. 

- En la segunda línea, que empezará por "A = ", se indican los símbolos del alfabeto de entrada. El conjunto de esos símbolos también irá comprendido entre llaves, y los símbolos irán separados por comas.

- La tercera línea, que comienza por "b" especificará el alfabeto de la pila. El conjunto de símbolos de la pila de nuevo irá comprendido entre llaves, y los símbolos estarán separados por comas. 

- La cuarta línea contiene el estado inicial del autómata. Dicha línea empieza por "q0 = " Este estado inicial irá entre llaves. 

- De forma análoga, la quinta línea, que comienza por "Z_0" contiene el símbolo inicial de la pila. El símbolo inicial también irá entre llaves.

- En la sexta línea se especifica el conjunto de estados finales del autómata. Tal línea empezará por "F = ". Igual que conjuntos anteriores, el conjunto de estados finales irá comprendido entre llaves y se utilizarán comas para separar cada uno de los estados.  

- A partir de la octava línea, tras una línea en blanco, se especifican las transiciones. Hau una línea por cada transición.  El formato de las transiciones es $(q,a,B) \rightarrow \{(p_1,B_1),\ldots,(p_n,B_n)\}$, donde $q$ es estado donde se encuentra el autómata antes de la transición, $a$ es el símbolo del alfabeto de entrada, $B$ es el tope de la pila y $(p_i, B_i)$ son cada uno de los pares (estado y nuevo tope de pila) a los que cambia el autómata desde $q$, con $B$ en el tope de la pila, leyendo $a, \quad \forall i = 1,2,\ldots,n$. 

Una vez creado este fichero, hay que indicar la ruta relativa donde se encuentra. Después se ha de llamar al método de clase readAutomaton pasándo como parámetro dicha ruta para generar el autómata. 


```python
path_automaton_stack = "automaton_stack.txt"
automaton_stack = AutomatonStack.readAutomaton(path_automaton_stack)
```

Del mismo modo, dado un objeto de la clase AutomatonStack, podemos escribir el autómata con pila en un fichero especificando la 
ruta de dicho fichero. Para ello, hemos emplear el método writeAutomaton. El formato de escritura de un autómata con pila es el mismo que el de lectura

```python
path_write_automaton_stack = "automaton_stack_written.txt"
automaton_stack.writeAutomaton(path_write_automaton_stack)
```

# Pertenencia de una palabra al lenguaje generado por el autómata con pila

Para comprobar la pertenencia de una palabra $u$ al lenguaje aceptado por un autómata con pila hemos de comprobar dicha pertenencia de forma recursiva partiendo desde la configuración inicial formada por el estado inicial, $u$ y el contenido inicial de la pila, que sería sólo el símbolo inicial de la misma. 

**Condición Base**: Si tenemos $(p, \epsilon, B)$ la palabra es aceptada sí, y sólo sí, $p$ es un estado final (criterio de estados finales), o $B = \epsilon$ (criterio de pila vacía)

**Condición recursiva**:    Si tenemos una configuración $(p, au, B)$, podemos pasar a una configuración $(q, u, B')$, en un paso de cálculo si $(q,B')\in \delta(p,a,B)$, donde $p \in Q$, $a \in A$, y $B$ y $B'$ son cadenas del alfabeto de la pila. Si para alguna de estas configuraciones a las que se puede llegar en un paso de cálculo $u$ es aceptada, entonces $au$ es también aceptada para $(p, au, B)$automata. Si $u$ no es aceptada para ninguna de estas configuraciones, entonces $u$ tampoco es aceptada para $(p, au, B)$. 

El método *checkBelonging* permite comprobar la pertenencia de una palabra al lenguaje generado por el autómata. 

    

```python
word = "00100"
belonging = automaton_stack.checkBelonging(word)
print(belonging)
```

# Autómatas con pila deterministas

Un autómata con pila se dice que es **determinista** si de una configuración puedo llegar, como mucho, a otra configuración en un paso de cálculo. Esto ocurre sí, y sólo sí, se cumplen las dos siguientes condiciones:

1. $\delta(q,a,X)$ tiene, como mucho, un elemento, $\quad \forall q \in Q, \quad a \in A \cup \{\epsilon\}, \quad X \in B$.

2. Si $\delta(q,a,X) \neq \emptyset $ para algún $a \in A$, entonces $\delta(q,\epsilon,X) = \emptyset, \quad \forall q \in Q,  \quad X \in B$.

El método  *isDeterministic* de la clase AutomatonStack comprueba si un autómata con pila es determinista. 

```python
deterministic = automaton_stack.isDeterministic()
print(deterministic)
```

# Equivalencia de estados de aceptación

Sabemos que, dado un autómata con pila que acepta un lenguaje por el criterio de pila vacía, existe otro autómata con pila que acepta el mismo lenguaje por el criterio de estados finales y viceversa.

## De pila vacía a estados finales. 

Supongamos que $M = (Q,A,B,\delta,q_0,Z_0,F)$ es un autómata con pila que acepta un lenguaje $L(M)$ por el criterio de pila vacía. El autómata $M_f$ que acepta $L(M)$ por el criterio de estados finales se contruye a partir de $M$ mediante los tres siguientes pasos:

1. Se añaden dos nuevos estados a $Q$: $q_0^{n}$ y $q_f$. $q_0^{n}$ será el estado inicial de $M_f$ y $q_f$ será nuevo estado final de $M_f$.

2. Se añade un nuevo símbolo a $B$: $Z_0^{n}$. Éste será el símbolo inicial de la pila en $M_f$.

3. Las transiciones de $M_f$ serán las transiciones de $M$ más las siguientes:
    
    - $\delta(q_0^{n}, \epsilon, Z_0^{n}) = \{(q_0, Z_0Z_0^{n})\}$
    
    - $\delta(q, \epsilon, Z_0^{n}) = \{(q_f, Z_0Z_0^{n})\}, \quad \forall q \in Q$

El método *equivalentAutomatonFinalStates* de la clase AutomatonStack devuelve un autómata que acepta el mismo lenguaje que el original por el criterio de estados finales. 


```python
path_automaton_final_states = "automaton_stack_final_states.txt"
automaton_final_states = automaton_stack.equivalentAutomatonFinalStates()
automaton_final_states.writeAutomaton(path_automaton_final_states)
```

## De estados finales a pila vacía

Sea $M = (Q,A,B,\delta,q_0,Z_0,F)$ es un autómata con pila que acepta un lenguaje $L(M)$ por el criterio de estados finales. El autómata $M_n$  que acepta $L(M)$ por el criterio de pila vacía se contruye a partir de $M$ mediante los tres siguientes pasos:

1. Se añaden dos nuevos estados a $Q$: $q_0^{n}$ y $q_s$. $q_0^{n}$ será el estado inicial de $M_n$.

2. Se añade un nuevo símbolo a $B$: $Z_0^{n}$. Éste será el símbolo inicial de la pila en $M_n$.

3. Las transiciones de $M_f$ serán las transiciones de $M$ más las siguientes:
    
    - $\delta(q_0^{n}, \epsilon, Z_0^{n}) = \{(q_0, Z_0Z_0^{n})\}$.
    
    - $\delta(q, \epsilon, H) = \{(q_s, H)\}, \quad \forall q \in F, \quad H \in B \cup \{Z_0^{n}\}$.
    
    - $\delta(q_s, \epsilon, H) = \{(q_s, H)\}, \quad H \in B \cup \{Z_0^{n}\}$


El método *equivalentAutomatonEmptyStack* de la clase AutomatonStack devuelve un autómata que acepta el mismo lenguaje que el original por el criterio de estados finales. 


```python
path_automaton_stack3 = "automaton_stack3.txt"
automaton_stack3 = AutomatonStack.readAutomaton(path_automaton_stack3)
automaton_empty_stack = automaton_stack3.equivalentAutomatonEmptyStack()
path_automaton_empty_stack = "automaton_empty_stack.txt"
automaton_empty_stack.writeAutomaton(path_automaton_empty_stack)
```

# Autómatas con pila y gramáticas independientes del contexto

## De gramática independiente del contexto a autómata con pila

Dada una gramática independiente del contexto $G = (V,T,P,S)$, existe un autómata con pila $M = (Q,A,B,\delta,q_0,Z_0,F)$ que acepta por el criterio de pila vacía el mismo lenguaje que el que genera $G$. Los elementos de dicho autómata son los siguientes:

- $Q = \{q\}$

- $A = T$

- $B = V \cup T$

- $F = \emptyset$

- $q_0 = q$

- $Z_0 = S$

- $\delta$ viene determinada por las siguientes transiciones:

    - $\delta(q,\epsilon,B) = \{(q,\alpha) \mid B \rightarrow \alpha \in P\}$
    - $\delta(q,a,a) = (q,\epsilon), \quad \forall a \in A$
    
La función *automatonGrammar* de la librería *automatonStack_ICGrammar* nos permite pasar una gramática independiente del contexto a un autómata con pila según el método que acabamos de describir.  

```python
from AutomatonStack_ICGrammar import automatonGrammar
from grammar import GenerativeGrammar

path_grammar_automaton_stack = "grammar_to_automaton_stack.txt"
grammar_to_automaton_stack = GenerativeGrammar.readGrammar(path_grammar_automaton_stack)
automaton_from_grammar = automatonGrammar(grammar_to_automaton_stack)

path_write_automaton = "automaton_from_grammmar.txt"
automaton_from_grammar.writeAutomaton(path_write_automaton)
```

## De autómata con pila a gramática independiente del contexto

Supongamos que tenemos un autómata con pila $M = (Q,A,B,\delta,q_0,Z_0,\emptyset)$ que acepta el lenguaje por el criterio de pila vacía. Entonces existe una gramática independiente del contexto $G = (V,T,P,S)$ que genera el lenguaje aceptado por $M$. Los elementos de tal gramática serán los siguientes.

- Las variables $V$ serán todas las variables de la forma $[q,C,p]$, donde $p, q \in Q, \quad C \in B$, además de una nueva variable $S$ que será la variable inicial.
- El conjunto de símbolos terminales $T$ coincide con $A$, el alfabeto de entrada del autómata.
- Para cada transición $(p, D_1, D_2, \ldots,D_m) \in \delta(q,a,C)$, se añaden todas las producciones de la forma:

$[q,C,q_m] \rightarrow a[p,D_1,q_1][q_1,D_2,q_2]\cdots[q_{m-1},D_m,q_m], \quad \forall q_1,q_2,\ldots,q_m \in Q, \quad a \in A \cup \{\epsilon\}$.

La función *grammarAutomatonStack* de la librería *automatonStack_ICGrammar* nos permite pasar un autómata con pila a gramática independiente del contexto en base a este método.  

```python
from AutomatonStack_ICGrammar import grammarAutomatonStack

path_automaton_stack_to_grammar = "automaton_stack_to_grammar.txt"
automaton_stack_to_grammar = AutomatonStack.readAutomaton(path_automaton_stack_to_grammar)
grammar_from_automaton_stack = grammarAutomatonStack(automaton_stack_to_grammar)

path_write_grammar_from_automaton_stack = "grammar_from_automaton_stack.txt"
grammar_from_automaton_stack.writeGrammar(path_write_grammar_from_automaton_stack)
```
